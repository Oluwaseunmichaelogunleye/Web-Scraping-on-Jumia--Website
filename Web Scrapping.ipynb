```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import requests
from bs4 import BeautifulSoup as bs
```

```python
# StepS: inspect the website you want to scrape
```

```python
# send http request to the web server. this allows us to fetch webpage html content
url = "https://www.jumia.com.ng/catalog/?q=shoe"
access = requests.get(url) 
```

```python
# display the content from the web pag
access.content
```

```python
# to parse the html content from the jumia page; to navigate and search the the parsed content
# to parse means to analyze

soup = bs(access.content)
soup
```

```python
# find is used to return single item from the web page
r = soup.find("div", {"class": "info"})
r
```

```python
# getting the name of an item
item_name = r.find("h3", {"class":"name"}).get_text(strip=True)
```

```python
# getting the new price

item_new_price = r.find("div", {"class":"prc"}).get_text(strip=True)
```

```python
# getting the old price

item_old_price = r.find("div", {"class":"old"}).get_text(strip=True)
```

```python
# number of people that rated the item

no_rated_ppl = r.find("div", {"class":"rev"}).text.split("(")[1].strip(")")
```

```python
# star rating og thr item

rated_score = r.find("div", {"class":"stars _s"}).text.split()[0]
```

```python
# discount percentage

discount_cent = r.find("div", {"class":"bdg _dsct _sm"}).text
```

```python
all_dic = {"item_name": item_name, "item_new_price": item_new_price, "item_old_price": item_old_price, "no_rated_ppl": no_rated_ppl, "rated_score": rated_score, "discount_cent": discount_cent  }
```

```python
all_dic
```

```python
pd.DataFrame(all_dic, index=[0])
```

```python
# findall will return the list of all items on web page

soup.findAll("div", {"class":"info"})
```

```python
len(soup.findAll("div", {"class":"info"}))
```

```python
# getting all the names

all_items = soup.findAll("div", {"class":"info"})
for i in all_items:
    try:
        print(i.find("h3", {"class":"name"}).get_text(strip=True))
    except:
        print(np.nan)  #np.nan is how to define a missing value
```

```python
#getting all the new prices

all_items = soup.findAll("div", {"class":"info"})
for i in all_items:
    try:
        print(i.find("div", {"class":"prc"}).get_text(strip=True))
    except:
        print(np.nan)
```

```python
# getting at the old prices
all_items = soup.findAll("div", {"class":"info"})
for i in all_items:
    try:
        print(i.find("div", {"class":"old"}).get_text(strip=True))
    except:
        print(np.nan)
```
